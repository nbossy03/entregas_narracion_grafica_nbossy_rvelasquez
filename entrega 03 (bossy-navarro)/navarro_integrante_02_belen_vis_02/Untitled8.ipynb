```python
import pandas as pd

# Read the CSV with the correct delimiter
df = pd.read_csv('/content/estaodos-dragofi.csv', delimiter=';')
df

import altair as alt
import pandas as pd

# Activa el renderizador adecuado
alt.renderers.enable('default')  # O usa 'colab' si 'default' no sirve

# Asegúrate de que los nombres de columna sean correctos
# This line should now work correctly as the columns should be properly parsed
print(df.columns)
print(df[['Estados', 'Participantes']].dropna().head())

# Crea el gráfico
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Estados:N', title='Estados'),
    y=alt.Y('Participantes:Q', title='Participantes'),
    tooltip=[
        alt.Tooltip('Participantes:Q', title='Participantes'),
        alt.Tooltip('Estados:N', title='Estado')
    ]
).properties(
    width=600,
    height=400,
    title='Cantidad de participantes por estado'
).interactive()

chart


```
